# 11.Создаем новый столбец total_income_category

```python
income_cat=len(df)*[0]
```

```python
for i in range(len(df)):
    if df['total_income'].loc[df.index[i]]<30000:
        income_cat[i]='E'
    elif df['total_income'].loc[df.index[i]]<50000:
         income_cat[i]='D'
    elif df['total_income'].loc[df.index[i]]<200000:
         income_cat[i]='C'
    elif df['total_income'].loc[df.index[i]]<1000000:
         income_cat[i]='B'
    else: 
         income_cat[i]='A'
```

```python
df.insert(9,'total_income_category', income_cat)
```

# 12.Создаем новый столбец purpose_cat

```python
def purpose_cat(list):
    if 'автомоб' in list:
        return "автомобиль"
    if "образ" in list:
        return "образование"
    if "свадьб" in list:
        return "свадьба"
    if "недвиж" in list:
        return "недвижимость"
    if "строит" in list:
        return "строительство"
    if "жил" in list:
        return "жилье"

df['purpose_cat'] = df['purpose'].apply(purpose_cat)
```
